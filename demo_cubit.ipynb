# Demostración del simulador de cúbit

Este cuaderno muestra:
- Cómo agregar un estado válido a un cúbit.
- El caso de un estado **no válido**.
- Cómo usar cada compuerta cuántica del simulador.

En cada caso, se muestra la esfera de Bloch resultante.

## Importación del simulador

```python
import numpy as np
from quantum.implementations.simulated_gates import SimulatedGates
from quantum.circuits.circuit import Circuit
from quantum.plotter.simulated_bloch_plotter import plot_qubit_on_bloch_sphere
```

## Estado válido

```python
# Estado |ψ> = (1/√2)|0> + (1/√2)|1>
state_valido = np.array([1/np.sqrt(2), 1/np.sqrt(2)], dtype=complex)
c = Circuit(1)
c.set_state(0, state_valido)
plot_qubit_on_bloch_sphere(state_valido)
```

## Estado no válido (no normalizado)

```python
state_invalido = np.array([2, 3], dtype=complex)
try:
    c.set_state(0, state_invalido)
except Exception as e:
    print("Error detectado:", e)
```

## Aplicación de compuertas cuánticas

## X

```python
sg = SimulatedGates()
qc = Circuit(1)
state = np.array([1.0, 0.0], dtype=complex)
qc.set_state(0, state)
qc.apply_operation(0, sg.X)
new_state = qc.get_qubit_state(0)
print(new_state)
plot_qubit_on_bloch_sphere(new_state)
```
